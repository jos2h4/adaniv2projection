```python
import numpy as np
import pandas as pd
import yfinance as yf
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
from google.colab import files

# Set random seed for reproducibility
np.random.seed(42)

# Parameters
TICKER = 'ADANIPORTS.NS'
LOOK_BACK = 60  # Number of previous days to use for prediction
EPOCHS = 50
BATCH_SIZE = 32

# Set date range for the most recent data
end_date = datetime.now()  # Get the current date
start_date = end_date - timedelta(days=30)  # Get the date 30 days before

print(f"Fetching data from {start_date.date()} to {end_date.date()}")

# Fetch stock data
print("Fetching stock data...")
try:
    # First, let's get historical data for training (2 years before start_date)
    historical_start = datetime(2023, 1, 1)  # 2 years of historical data
    stock_data = yf.download(TICKER, start=historical_start, end=end_date)

    if stock_data.empty:
        raise Exception("No data fetched")

    print(f"\nData range: from {stock_data.index.min()} to {stock_data.index.max()}")
    print(f"Total trading days: {len(stock_data)}")

    # Display the most recent month of data
    recent_data = stock_data['2025-02-01':]
    print("\nMost recent data (2025):")
    display(recent_data.head())

except Exception as e:
    print(f"Error fetching data: {e}")
    raise

# Display basic statistics of recent data
print("\nBasic statistics of the recent stock data (2025):")
display(recent_data.describe())

# Prepare data
print("\nPreparing data for LSTM...")
dataset = stock_data['Close'].values.reshape(-1, 1)
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(dataset)

# Create sequences for training
X, y = [], []
for i in range(LOOK_BACK, len(scaled_data)):
    X.append(scaled_data[i-LOOK_BACK:i, 0])
    y.append(scaled_data[i, 0])

X, y = np.array(X), np.array(y)
X = np.reshape(X, (X.shape[0], X.shape[1], 1))

# Split into train and test sets (80-20 split)
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

print(f"\nTraining data size: {len(X_train)} days")
print(f"Testing data size: {len(X_test)} days")

# Create and compile LSTM model
print("\nCreating and training LSTM model...")
model = Sequential([
    LSTM(units=50, return_sequences=True, input_shape=(LOOK_BACK, 1)),
    Dropout(0.2),
    LSTM(units=50, return_sequences=False),
    Dropout(0.2),
    Dense(units=25),
    Dense(units=1)
])

model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model with progress bar
history = model.fit(
    X_train, y_train,
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    validation_split=0.1,
    verbose=1
)

# Make predictions
print("\nMaking predictions...")
train_predict = model.predict(X_train)
test_predict = model.predict(X_test)

# Inverse transform predictions
train_predict = scaler.inverse_transform(train_predict)
test_predict = scaler.inverse_transform(test_predict)
y_train_actual = scaler.inverse_transform([y_train])
y_test_actual = scaler.inverse_transform([y_test])

# Calculate RMSE
train_rmse = np.sqrt(np.mean((train_predict - y_train_actual.T) ** 2))
test_rmse = np.sqrt(np.mean((test_predict - y_test_actual.T) ** 2))
print(f'\nTrain RMSE: {train_rmse:.2f}')
print(f'Test RMSE: {test_rmse:.2f}')

# Plot predictions (focus on 2025 data and future predictions)
plt.figure(figsize=(16, 8))

# Plot actual prices (from recent_data)
plt.plot(recent_data.index, recent_data['Close'], label='Actual Price', color='blue')

# Plot predicted prices (from test_predict)
plt.plot(recent_data.index, test_predict[-len(recent_data):], label='Predicted Price (Past)', color='red', linestyle='--')

plt.title(f'{TICKER} Stock Price Prediction')
plt.xlabel('Date')
plt.ylabel('Stock Price (INR)')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot training history
plt.figure(figsize=(16, 8))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.tight_layout()
plt.show()

# Save predictions to CSV and download in Colab
recent_actual = y_test_actual[0][-len(recent_data):]
recent_pred = test_predict[-len(recent_data):]

results_df = pd.DataFrame({
    'Date': recent_data.index,
    'Actual': recent_actual.flatten(),
    'Predicted': recent_pred.flatten()
})
results_df.to_csv('ADANIPORTS_predictions_2025.csv')
files.download('ADANIPORTS_predictions_2025.csv')

# Print model summary
print("\nModel Summary:")
model.summary()

# Display final evaluation metrics
print("\nFinal Evaluation Metrics:")
print(f"Training RMSE: {train_rmse:.2f} INR")
print(f"Testing RMSE: {test_rmse:.2f} INR")

# Display recent predictions
print("\nMost Recent Predictions (2025):")
display(results_df.tail())
```
